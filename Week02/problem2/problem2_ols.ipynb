```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
```

```python
#import data
df = pd.read_csv('problem2.csv')
x = df['x'].values
y = df['y'].values
```

```python
#calculation
#mean
mean_x = np.mean(x)
mean_y = np.mean(y)
#slop m
numer = 0
denom = 0
for i in range(0, len(x)):
    numer += (x[i]-mean_x) * (y[i]-mean_y)
    denom += (x[i]-mean_x) ** 2
m = numer/denom
c = mean_y - m * mean_x
print("the slope is " + str(m))
print("the intercept is " + str(c))
```

```python
max_x = np.max(x)
min_x = np.min(x)

x2 = np.linspace (min_x, max_x, 100)
y2 = c + m * x2

plt.plot(x2, y2, color='green', label='Regression Line')
plt.scatter(x, y, c='orange', label='data points')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
error = y - m * x - c
df_err = pd.DataFrame(error)
```

```python
#drawing pictures
fig = plt.figure(figsize = (10,6))
ax1 = fig.add_subplot(2,1,1)
ax1.scatter(df_err.index, df_err.values)
plt.grid()

ax2 = fig.add_subplot(2,1,2)
df_err.hist(bins=30,alpha = 0.5,ax = ax2)
df_err.plot(kind = 'kde', secondary_y=True,ax = ax2)
plt.grid()
```

```python
#ks test
u = df_err.mean()
std = df_err.std()
stats.kstest(df_err, 'norm', (u, std))
```
