```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#import data
df = pd.read_csv('problem1.csv')
x = df['x'].values
y = df['y'].values

# calculate the mean of X,Y
mean_x = np.mean(x)
mean_y = np.mean(y)

sigMat = np.cov(x,y)
# calculation using conditional multivariate normal distribution
y_hat = mean_y+sigMat[0][1]/sigMat[0][0]*(x-mean_x)

ConVar_y = sigMat[1][1]-sigMat[0][1]*sigMat[0][1]/sigMat[0][0]
print(ConVar_y)
```

```python
#calculation by ols
#mean
mean_x = np.mean(x)
mean_y = np.mean(y)
#slop m
numer = 0
denom = 0
for i in range(0, len(x)):
    numer += (x[i]-mean_x) * (y[i]-mean_y)
    denom += (x[i]-mean_x) ** 2
m = numer/denom
c = mean_y - m * mean_x
error = y - m * x - c
print(np.var(error))
```

```python
diff = y_hat - (m * x + c)
plt.plot(diff)
```
